```python
import numpy as np
```

```python
datalines = open(r"C:\Users\nothi\Documents\Studies\Thesis\datasets\Temporal graphs\email-dnc\email-dnc.edges", 'r').readlines()
print(datalines[:10])
datalines = [row.replace('\n', '').split(',') for row in datalines]
```

```python
froms, tos, times = zip(*datalines)
```

```python
int_froms = list(map(int, froms))
int_tos = list(map(int, tos))
int_times = list(map(int, times))
```

```python
DAY_IN_SECONDS = 60*60*24
WEEK_IN_SECONDS = DAYS_IN_SECONDS*7
```

```python
min_time = np.min(int_times)
max_time = np.max(int_times)
time_span_seconds = max_time-min_time
num_of_days = int(np.ceil(time_span_seconds/DAY_IN_SECONDS))
num_of_weeks = int(np.ceil(time_span_seconds/WEEK_IN_SECONDS))
```

```python
print(f"num of days: {num_of_days}, num of weeks: {num_of_weeks}".format())
```

```python
np.histogram(int_times, bins=num_of_weeks)
```

```python
days = []
for i in range(num_of_days):
    day_start = min_time + i*(DAY_IN_SECONDS)
    day_end = np.min([max_time, min_time + (i+1)*(DAY_IN_SECONDS)])
    days.append((day_start, day_end))
```

```python
weeks = []
for i in range(num_of_weeks):
    week_start = min_time + i*(WEEK_IN_SECONDS)
    week_end = np.min([max_time, min_time + (i+1)*(WEEK_IN_SECONDS)])
    weeks.append((week_start, week_end))
```

```python
print(f"num of days: {len(days)}, num of weeks: {len(weeks)}".format())
```

```python
graphs_split_to_days = [[] for i in range(num_of_days)]
for n1, n2, t in zip(int_froms, int_tos, int_times):
#     print(type(n1), type(n2), type(t))
    day_egde_belongs_to = int(np.floor((t - min_time) / DAY_IN_SECONDS))
    if (n1 == n2):
        continue
    graphs_split_to_days[day_egde_belongs_to].append((n1, n2))
```

```python
graphs_split_to_weeks = [[] for i in range(num_of_weeks)]
for n1, n2, t in zip(int_froms, int_tos, int_times):
#     print(type(n1), type(n2), type(t))
    week_egde_belongs_to = int(np.floor((t - min_time) / WEEK_IN_SECONDS))
    if (n1 == n2):
        continue
    graphs_split_to_weeks[week_egde_belongs_to].append((n1, n2))
```

```python
print('day graphs num of edges:')
for g in graphs_split_to_days:
    print(len(g))

print('week graphs num of edges:')
for g in graphs_split_to_weeks:
    print(len(g))
```

```python
import networkx as nx
```

```python
nx_graphs_split_to_days = list(map(nx.DiGraph, graphs_split_to_days))
nx_graphs_split_to_weeks = list(map(nx.DiGraph, graphs_split_to_weeks))
```

```python
num_of_nodes_per_day = [g.number_of_nodes() for g in nx_graphs_split_to_days]
num_of_nodes_per_week = [g.number_of_nodes() for g in nx_graphs_split_to_weeks]

num_of_edges_per_day = [g.number_of_edges() for g in nx_graphs_split_to_days]
num_of_edges_per_week = [g.number_of_edges() for g in nx_graphs_split_to_weeks]
```

```python
weekly_graphs_one_with_lables = [{'graph': g} for g in interesting_weekly_graphs_candidate_one]
weekly_graphs_two_with_lables = [{'graph': g} for g in interesting_weekly_graphs_candidate_two]
```
